```python
#Matplotlib exercises
Df= pd.read_csv('company_sales_data.csv')
```

```python
Df.head(3)
```

```python
Df.isnull().values.any()
```

```python
Df.dtypes
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
Df.plot('month_number','total_profit',color='blue')
plt.xlabel('month number')
plt.ylabel('Total profit')
plt.title('Company profit per month')
plt.show()
```

```python
ax=plt.gca()
Df.plot('month_number','facecream',color='red',label='facecream',ax=ax)
Df.plot('month_number','facewash',color='green',label='facewash',ax=ax)
Df.plot('month_number','toothpaste',color='blue',label='toothpaste',ax=ax)
Df.plot('month_number','bathingsoap',color='black',label='toothpaste',ax=ax)
Df.plot('month_number','shampoo',label='shampoo',ax=ax)
Df.plot('month_number','moisturizer',label='moisturizer',ax=ax)
plt.ylabel('sales unit in number')
plt.legend()
plt.show()
```

```python
Df.plot(kind='scatter',x='month_number',y='toothpaste',color='red')
plt.ylabel('Units sold')
plt.title('Toothpaste sales data')
plt.show()
```

```python

fig, ax = plt.subplots()
monthList  = Df ['month_number'].tolist()
facecreamm=Df['facecream'].tolist()
facewashh=Df['facewash'].tolist()
plt.bar([a-0.25 for a in monthList],facecreamm, color='red',width= 0.25,align='edge')
plt.bar([a+0.25 for a in monthList],facewashh,  color='blue',width= -0.25,align='edge')
plt.ylabel('Units sold')
plt.title('Facewash and facecream sales data')
plt.xticks(monthList)
plt.grid(True, linewidth= 1, linestyle="--")
plt.legend()
plt.show()
```

```python
Df.tail(3)
```

```python
facecream_sum= Df['facecream'].sum()
facewash_sum=Df['facewash'].sum()
tooth_sum=Df['toothpaste'].sum()
bathing_sum=Df['bathingsoap'].sum()
shampoo_sum=Df['shampoo'].sum()
moisturizer_sum=Df['moisturizer'].sum()
salesdata=[facecream_sum,facewash_sum,tooth_sum,bathing_sum,shampoo_sum,moisturizer_sum]
```

```python
my_labels='facecream','facewash','toothpaste','bathingsoap','shampoo','moisturizer'
plt.pie(salesdata,labels=my_labels,autopct='%1.1f%%')
plt.legend(loc='lower right')
plt.axis('equal')
plt.show()
```
